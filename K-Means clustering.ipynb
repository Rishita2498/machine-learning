```python
l=[]
for  i in range(1,20):
    kmean=KMeans(n_clusters=i)
    kmean.fit(x)
    l.append(kmean.inertia_)
print(l)
```

```python
l
```

```python
plt.figure(figsize=(25,6))
plt.plot(l,'b')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.title('Elbow method')
plt.show
```
