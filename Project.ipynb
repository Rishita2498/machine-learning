## Step-2:Extracting independent and dependent variables

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

## Normalisation

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_sc=sc.fit_transform(x)
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(x_sc,vert=False,labels=['Refractive Index', 'Sodium', 'Magnesium', 'Aluminium', 'Silicon',
       'Potassium', 'Calcium', 'Barium', 'Iron'],patch_artist=True)
plt.title('z - score')
plt.show()
```

```python
x_sc
```

```python
x_sc.shape,y.shape
```

```python
y=y.reshape(-1,1)
```

### Splitting data into training and testing

```python
from sklearn import cross_validation,neighbors
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sc,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### Building a Machine Learning model
#### Logistic Regression,K - NN,SVM,Decision Tree,Random Forest

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model_log=LogisticRegression(C=20.0)
model_knn=KNeighborsClassifier(n_neighbors=3)
model_svm=SVC(kernel='linear')
model_dt=DecisionTreeClassifier()
model_rf=RandomForestClassifier(n_estimators=25)
```

```python
model_log.fit(x_train,y_train)
model_knn.fit(x_train,y_train)
model_svm.fit(x_train,y_train)
model_dt.fit(x_train,y_train)
model_rf.fit(x_train,y_train)
print("Model trained Successfully!!")
```

## Model Evaluation
#### - Confusion Matrix
#### - Classification report


```python
y_log=model_log.predict(x_test)
y_knn=model_knn.predict(x_test)
y_svm=model_svm.predict(x_test)
y_dt=model_dt.predict(x_test)
y_rf=model_rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm_log=confusion_matrix(y_test,y_log)
cm_knn=confusion_matrix(y_test,y_knn)
cm_svm=confusion_matrix(y_test,y_svm)
cm_dt=confusion_matrix(y_test,y_dt)
cm_rf=confusion_matrix(y_test,y_rf)

cr_log=classification_report(y_test,y_log)
cr_knn=classification_report(y_test,y_knn)
cr_svm=classification_report(y_test,y_svm)
cr_dt=classification_report(y_test,y_dt)
cr_rf=classification_report(y_test,y_rf)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(cm_log,annot=True,cbar=False,cmap='summer')
plt.xlabel('True Values'),plt.ylabel('Predicted Values')
plt.title("Logistic Regression")
plt.show()

plt.figure(figsize=(4,4))
sns.heatmap(cm_knn,annot=True,cbar=False,cmap='Paired')
plt.xlabel('True Values'),plt.ylabel('Predicted Values')
plt.title("K Nearest Neighbour")
plt.show()

plt.figure(figsize=(4,4))
sns.heatmap(cm_svm,annot=True,cbar=False,cmap='gnuplot')
plt.xlabel('True Values'),plt.ylabel('Predicted Values')
plt.title("Support vector Machine")
plt.show()

plt.figure(figsize=(4,4))
sns.heatmap(cm_dt,annot=True,cbar=False,cmap='rainbow')
plt.xlabel('True Values'),plt.ylabel('Predicted Values')
plt.title("Decision Tree")
plt.show()

plt.figure(figsize=(4,4))
sns.heatmap(cm_rf,annot=True,cbar=False,cmap='gnuplot2')
plt.xlabel('True Values'),plt.ylabel('Predicted Values')
plt.title("Random Forest")
plt.show()
```

```python
print('='*20+'Logistic Regression'+'='*20+'\n')
print(cr_log)
print('='*20+'K Nearest Neighbour'+'='*20+'\n')
print(cr_knn)
print('='*20+'Support vector Machine'+'='*20+'\n')
print(cr_svm)
print('='*20+'Decision Tree'+'='*20+'\n')
print(cr_dt)
print('='*20+'Random Forest'+'='*20+'\n')
print(cr_rf)
```
